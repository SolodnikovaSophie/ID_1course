```python
from bs4 import BeautifulSoup
import numpy as np
import json
import re

def handle_file(path):
    with open(path, 'r', encoding="utf-8") as f:
        xml_file = f.read()
        
        clothings = BeautifulSoup(xml_file, 'xml').find_all('clothing')

        item = dict()
        for clothing in clothings:
            item["id"] = int(clothing.id.get_text())
            item["name"] = clothing.find_all('name')[0].get_text().strip()
            item["category"] = clothing.category.get_text().strip()
            item['size'] = clothing.size.get_text().strip()
            item['color'] = clothing.color.get_text().strip()
            item['material'] = clothing.material.get_text().strip()
            item['price'] = float(clothing.price.get_text().strip())
            item['rating'] = float(clothing.rating.get_text().strip())
            item['reviews'] = int(clothing.reviews.get_text().strip())
            if clothing.sporty is not None:
                item['sporty'] = clothing.sporty.get_text().strip() == "yes"
            if clothing.new is not None:
                item['new'] = clothing.new.get_text().strip() == "+"
            if clothing.exclusive is not None:
                item["exclusive"] = clothing.exclusive.get_text().strip() == "yes"
            items.append(item)
    return item

items = []
for i in range(1, 180):
    file_name = rf"C:\Users\ipavl\Учеба_УРФУ\1_курс\1_семестр\ИД\Practice_3\86\4\{i}.xml"
    result = handle_file(file_name)
    items.append(result)

with open('result_4.json', 'w', encoding="utf-8") as f:
    json.dump(items, f, ensure_ascii=False)

print(items)
```

```python
items_filtered = sorted(items, key = lambda x: x['rating'], reverse = True)
print(items_filtered)

with open('result_4_filtered.json', 'w', encoding="utf-8") as f:
    json.dump(items_filtered, f, ensure_ascii=False)
```

```python
material_filt = [item for item in items if (item['material'] == 'Лен') & (item['color'] == 'Синий')]
print(material_filt)

with open('result_4_filtered_1.json', 'w', encoding="utf-8") as f:
    json.dump(material_filt, f, ensure_ascii=False)
```

```python
price_stat = np.array([item['price'] for item in items])

price_stats = {
    "Максимальное значение": int(np.max(price_stat)),
    "Минимальное значение": int(np.min(price_stat)),
    "Сумма значений": int(np.sum(price_stat)),
    "Среднее арифметическое значение": int(np.mean(price_stat))
}

with open('price_stats_2.json', 'w', encoding="utf-8") as f:
    json.dump(price_stats, f, ensure_ascii=False)

print(price_stats)
```

```python
title_freq = np.array([item['color'] for item in items])

word_freq = {}
for word in title_freq:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

with open('word_freq_4.json', 'w', encoding="utf-8") as f:
    json.dump(word_freq, f, ensure_ascii=False)

print(word_freq)
```
